```python
from nfinder import *
```

```python
points = 3*np.array([[0, 1], [2, 3], [2, 2], [1, 1], [1,0]])

df_edges, thresholds = edgeprops(points)
edges = neighbors_graph(points)
```

```python
titlesize = 18

fig, axes = plt.subplots(1, 3, figsize=(3*5, 5))

axes[0].set_title('Sample', size=titlesize)
axes[0].plot(points[:, 0], points[:, 1], 'o')

axes[1].set_title('Delaunay graph', size=titlesize)
for source, target, isNeighbor in np.array(df_edges[['source', 'target', 'isNeighbor']]):
  if isNeighbor:
    color='black'
  else:
    color='red'
  axes[1].plot([points[source][0], points[target][0]], [points[source][1], points[target][1]], 'o', ls='-', 
               color=color)

axes[2].set_title('Neighbors graph', size=titlesize)
for source, target in edges:
  axes[2].plot([source[0], target[0]], [source[1], target[1]], 'o', ls='-', color='black')

plt.xlim([-1, 10])
plt.ylim([-1, 10])
for ax in axes:
  ax.set(xticks=[], yticks=[])

plt.tight_layout(pad=0)
fig.subplots_adjust(wspace=0.1)
plt.show()
```
